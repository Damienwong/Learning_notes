### isinstance 的作用
- 检查对象my_dict是否是指定类或其子类的实现；
- my_dict是dict类型，dict是Mapping的子类，因此返回Ture。

```python
from collections import abc
```

```python
my_dict = {}
isinstance(my_dict, abc.Mapping)
```

```python
type(my_dict)
```

## 字典的不同构建方式

```python
a = dict(one=1, two=2, three=3)
```

```python
a
```

```python
b = {'two': 2, 'one':1, 'three':3}
```

```python
b
```

```python
c = dict(zip(['one', 'two', 'three'], [1, 2, 3]))
```

```python
c
```

```python
d = dict([('one', 1), ('two', 2), ('three', 3)])
```

```python
d
```

```python
e = dict({'three': 3, 'two': 2, 'one': 1})
```

```python
e
```

```python
a == b == c == d == e
```

- 字典的推导

```python
items = [
    ('one', 1),
    ('two', 2),
    ('three', 3)
]

f = {english: digit for english, digit in items}
```

```python
f
```

```python
f.get('five', 4)
```

```python
f
```

### 集合

集合中的元素必须是可散列的。

```python
a = {1, 2, 3}
b = {2, 3, 4, 5}

print('合集：', a | b) 
print('交集：', a & b)
print('差集：', a - b)
```

```python
a.add(10)
```
