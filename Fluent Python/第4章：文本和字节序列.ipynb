## 字节    二进制序列

二进制序列其实是整数序列，各个字节的值可能会使用下列三种不同的方式显示：
- 可打印的ASCII范围内的字节(从空格到\~，即\x20~\x7E),使用ASCII字符本身；
- 部分特殊字节使用转义序列，如：制表符(\x09)为\t，换行符(\x0A)为\n，回车符(\x0D)为\r，\\(\x5C)为\\\；
- 其他字节的值，使用十六进制转义序列

```python
b'\x00\x48\x56\x0A\x5C'
```

```python
bytes.fromhex('314bcea9')
```

```python
a = b'\x48\x48\x48\x48\x48\x48\x48\x48\x48\x48'
```

```python
import sys, locale

expressions = '''
    locale.getpreferredencoding()
    sys.stdout.isatty()
    sys.stdout.encoding
    sys.stdin.isatty()
    sys.stdin.encoding
    sys.stderr.isatty()
    sys.stderr.encoding
    sys.getdefaultencoding()
    sys.getfilesystemencoding()
'''

for expression in expressions.split():
    value = eval(expression)
    print(expression.rjust(30), '->', repr(value))
```
